```julia
bc = loadtable(path, filenamecol=:Exchange)
```

```julia
g = [CountMap(String) Ignored() Hist(20) Hist(20) Hist(20) Hist(20) Hist(20) Hist(20) Hist(20)]

g = fit!(g, bc)
```
