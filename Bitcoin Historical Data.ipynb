```julia
addprocs()
using JuliaDB, OnlineStats, Plots
gr()
```

```julia
path = "/Users/joshday/datasets/bitcoin-historical-data"
```

```julia
bc = loadtable(path, filenamecol=:Exchange)
```

```julia
@everywhere f(x::DataValues.DataValue) = get(x)
@everywhere f(x::Real) = x
@everywhere f(x::NamedTuples.NamedTuple) = f.(x)


o = series(IndexedPartition(Int, Extrema()), 
    filter = x -> !any(isnull.(x)), transform = f)

out = collect(groupreduce(o, bc, :Exchange; select=(:Timestamp, :Weighted_Price)))
```

```julia
plotvec = [plot(out[i][2], title = out[i][1]) for i in 1:length(out)]
plot(plotvec..., layout = (7, 1), size = (800,1000), legend=false, linkx=true)
```
