# Exploring Continuous Data

Here we explore 5 million random observations

```julia
using OnlineStats, Plots
plotly()
y = randn(5_000_000);
```

```julia
o = Hist(100)  # adaptively make histogram of 100 bins
s = Series(o)

fit!(s, y)

plot(s)
```

- Now that the `Series` has run through the data vector, we can use the `Hist` object to get summary statistics.
    - `mean`, `var`, `std`, `median`, etc.

```julia
# approximate quantiles of the actual data
@time quantile(o)
```

```julia
# true quantiles.  How close are the approximate quantiles?
@time quantile(y)
```
